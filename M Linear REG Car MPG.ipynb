```python
data=data.drop('car_name',axis=1)
#droping car_name as it is not useful for the model
```

```python
data['origin']=data['origin'].replace({1:'america',2:'europe',3:'asia'})
```

```python
data=pd.get_dummies(data,columns=['origin'])
```

```python
#hp coloumn is not present which means their is some data impurity in it.
import numpy as np
```

```python
data=data.replace('?',np.nan)
```

```python
data=data.dropna()
```

```python
data['hp']=data['hp'].astype('float64')
```

# _____________________________________________________________________________________________________________________________

```python
X= data.drop('mpg',axis=1)
```

```python
y=data['mpg']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
Lreg= LinearRegression()
```

```python
Lreg.fit(X_train,y_train)
```

```python
for idx,col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {} ".format(col_name,Lreg.coef_[idx]))
    #coefficient of each independent variable
```

```python
Lreg.score(X_train,y_train)
```

```python
Lreg.score(X_test,y_test)
```

```python
y_test.reset_index()
```

```python
pred=Lreg.predict(X_test)
print(pred)
```
